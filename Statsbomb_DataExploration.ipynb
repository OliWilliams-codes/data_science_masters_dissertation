**Data construction and feature extraction**

First add any useful fields that can be universally useful such as x, y

x range is standardised 0 to 120 and y range is standardised 0 to 80

```python
#location data
df_final_match[['x','y']] = df_final_match['location'].apply(pd.Series)
```

Because of this structuring the data is essentially numerous small dataframes. We can get around this by breaking down our data into type dataframes

```python
#
new_dfs = {}
for event_type in df_final_match['type'].unique():
  df_name = f"df_{event_type.replace(' ', '_').lower()}"
  new_dfs[df_name] = df_final_match[df_final_match['type'] == event_type].copy()

print("List of new dataframes:")
for df_name in new_dfs.keys():
    print(df_name)
```

Use these new dataframes to reduce the data down to the appropriate fields.
A example will be shown below for dribble data

**df_dribble**

Can be seen that under_pressure, duration, and counterpress do not add any potential value due to having fixed values

```python
#select key columns
selected_columns = [

    'id',
    'index',
    'match_id',
    'period',
    'possession',
    'possession_team',
    'under_pressure',
    'duration',
    'dribble_outcome',
    'counterpress',
    'x',
    'y'
]
# Check if the columns exist in the database
df_dribble = new_dfs['df_dribble']

existing_columns = [col for col in selected_columns if col in df_dribble.columns]

# Create a new dataframe with only the existing selected columns
df_final_matches_selected_dribble = df_dribble[existing_columns]
df_final_matches_selected_dribble
```

Can then summarise the data by match. Using this method you can create, combine and analyse impactful match statistics for instance you could test factors in an xg model

```python
# Group by match_id
match_analysis_dribble = df_final_matches_selected_dribble.groupby('match_id').agg(
    dribble_count=('id', 'count'),
    dribble_proportion_period_1=('period', lambda x: (x == 1).mean()),
    dribble_success_rate=('dribble_outcome', lambda x: (x == 'Complete').mean()),
    dribble_start_mean=('x', 'mean'),
    dribble_start_proportion_x_lt_60=('x', lambda x: (x < 60).mean()), # Proportion of carry start x location when x is less than 60 e.g. in own half as pitch x axis is 0 to 120
    dribble_start_y_between_20_60_proportion=('y', lambda y: ((y >= 20) & (y <= 60)).mean()) # Proportion of carry start y location when y is between 20 and 60 i.e.
).reset_index()

match_analysis_dribble
```
